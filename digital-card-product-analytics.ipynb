# Merging similar datasets for a unified analysis

```python
# Create final analytical dataset 

bank_df = df.merge(
    df2,
    left_on=["card_id", "client_id"],
    right_on=["id", "client_id"],
    how="left",
    suffixes=("_transaction", "_card")
)

bank_df = bank_df.merge(
    df1,
    left_on="client_id",
    right_on="id",
    how="left",
    suffixes=("", "_user")
)

bank_df = bank_df.merge(
    df3,
    left_on="mcc",
    right_on="MCC_Code",
    how="left"
)

bank_df = bank_df.merge(
    df4,
    left_on="id_transaction",
    right_on="id",
    how="left",
    suffixes=("", "_fraud")
)

bank_df.head()
```

```python
print("Transactions rows:", df.shape[0])
print("Final dataset rows:", bank_df.shape[0])
print("Columns:", bank_df.shape[1])
```

### Interpretation

The final analytical dataset contains the same number of rows as the original transactions table, which means the merge did not duplicate or lose transaction records. This gives a reliable base for further product analysis.

```python
# Product analytics features

bank_df["is_purchase"] = bank_df["amount"] > 0
bank_df["is_refund"] = bank_df["amount"] < 0

# High-ticket threshold: top 20% of positive transactions
high_ticket_threshold = bank_df.loc[
    bank_df["is_purchase"], 
    "amount"
].quantile(0.80)

bank_df["is_high_ticket"] = (
    bank_df["is_purchase"] &
    (bank_df["amount"] > high_ticket_threshold)
)

# Amount as % of credit limit
bank_df["amount_to_limit_ratio"] = np.where(
    bank_df["credit_limit"] > 0,
    bank_df["amount"] / bank_df["credit_limit"],
    np.nan
)

# Potential installment opportunity:
# purchase amount is at least 10% of the card credit limit
bank_df["installment_opportunity"] = (
    bank_df["is_purchase"] &
    (bank_df["card_type"] == "Credit") &
    (bank_df["credit_limit"] > 0) &
    (bank_df["amount_to_limit_ratio"] >= 0.10)
)
# Fraud flag, depending on the original column name
bank_df["is_fraud"] = bank_df["Target"].map({
    "Yes": 1,
    "No": 0,
    True: 1,
    False: 0,
    1: 1,
    0: 0
})

high_ticket_threshold
```

# Cashback category scoring

```python
mcc_metrics = (
    bank_df[bank_df["is_purchase"]]
    .groupby(["mcc", "Description"], dropna=False)
    .agg(
        transactions_count=("id_transaction", "count"),
        unique_clients=("client_id", "nunique"),
        unique_cards=("card_id", "nunique"),
        total_volume=("amount", "sum"),
        avg_amount=("amount", "mean"),
        median_amount=("amount", "median"),
        high_ticket_share=("is_high_ticket", "mean"),
        installment_opportunity_share=("installment_opportunity", "mean"),
        fraud_rate=("is_fraud", "mean")
    )
    .reset_index()
)
```

```python
def min_max_scale(series):
    if series.max() == series.min():
        return series * 0
    return (series - series.min()) / (series.max() - series.min())

mcc_metrics["frequency_score"] = min_max_scale(mcc_metrics["transactions_count"])
mcc_metrics["reach_score"] = min_max_scale(mcc_metrics["unique_clients"])
mcc_metrics["volume_score"] = min_max_scale(mcc_metrics["total_volume"])

mcc_metrics["repeat_proxy"] = (
    mcc_metrics["transactions_count"] / mcc_metrics["unique_clients"]
)

mcc_metrics["repeat_proxy_score"] = min_max_scale(mcc_metrics["repeat_proxy"])
mcc_metrics["fraud_risk_score"] = mcc_metrics["fraud_rate"].fillna(0)

mcc_metrics["cashback_score"] = (
    0.30 * mcc_metrics["frequency_score"] +
    0.25 * mcc_metrics["reach_score"] +
    0.20 * mcc_metrics["repeat_proxy_score"] +
    0.15 * mcc_metrics["volume_score"] -
    0.10 * mcc_metrics["fraud_risk_score"]
)

cashback_recommendations = (
    mcc_metrics
    .sort_values("cashback_score", ascending=False)
    .head(10)
)

cashback_recommendations
```

### Interpretation

Everyday spending categories such as grocery stores, service stations, food stores, restaurants, and pharmacies appear to be strong cashback candidates. These categories combine high transaction frequency, broad customer reach, and relatively low fraud risk. This suggests that cashback category selection should not be based on transaction volume alone.

# Hypothesis testing

## Hypothesis 1. Credit cards have a higher share of high-ticket transactions than debit cards. 

```python
from statsmodels.stats.proportion import proportions_ztest

credit_debit = bank_df[
    (bank_df["is_purchase"]) &
    (bank_df["card_type"].isin(["Credit", "Debit"]))
].copy()

summary = credit_debit.groupby("card_type")["is_high_ticket"].agg(["sum", "count"])
summary
```

```python
count = summary.loc[["Credit", "Debit"], "sum"].values
nobs = summary.loc[["Credit", "Debit"], "count"].values

stat, p_value = proportions_ztest(
    count,
    nobs,
    alternative="larger"
)

print("z-statistic:", stat)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    print("Result: Credit cards have a statistically significantly higher share of high-ticket transactions than debit cards.")
else:
    print("Result: No statistically significant evidence that credit cards have a higher high-ticket share.")
```

### Interpretation

The proportion z-test confirms that credit cards have a statistically significantly higher share of high-ticket transactions than debit cards. The difference is also practically meaningful: around 28% of credit card purchases are high-ticket transactions, compared with around 18% for debit cards. This supports the idea that credit cards are more strongly associated with larger purchases.

## Hypothesis 2. Online transactions have a different fraud rate than offline transactions.

```python
from scipy.stats import chi2_contingency

fraud_channel = bank_df[
    bank_df["is_purchase"] &
    bank_df["is_fraud"].notna()
].copy()

fraud_channel["is_online"] = fraud_channel["use_chip"].eq("Online Transaction")

contingency_table = pd.crosstab(
    fraud_channel["is_online"],
    fraud_channel["is_fraud"]
)

contingency_table
```

```python
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print("chi-square:", chi2)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    print("Result: Fraud rate differs significantly between online and offline transactions.")
else:
    print("Result: No statistically significant evidence of fraud rate difference between online and offline transactions.")
```

### Interpretation

The chi-square test shows a statistically significant difference in fraud distribution between online and offline transactions. Online transactions have a noticeably higher fraud rate, which suggests that digital transaction channels require stronger risk monitoring when designing product campaigns or encouraging online card usage.
